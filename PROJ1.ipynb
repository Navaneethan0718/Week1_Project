```python
quantity = data['Quantity_Sold']
unit_price = data['Unit_Price']
sales = data['Sales_Amount']
```

```python
calculated_sales = quantity * unit_price
```

```python
print("Total Revenue:", np.sum(calculated_sales))
print("Average Sales:", np.mean(calculated_sales))
```

```python
print("Mean:", np.mean(sales))
print("Median:", np.median(sales))
print("Variance:", np.var(sales))
print("Standard Deviation:", np.std(sales))
```

```python
print("First 5 Sales Records:", sales[:5])
print("High Sales Records:", sales[sales > 5000])
```

```python
unit_cost = data['Unit_Cost']
profit = (unit_price - unit_cost) * quantity

print("Total Profit:", np.sum(profit))
print("Average Profit:", np.mean(profit))
```
